```python
import json
import pathlib
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

```

## Read data 

```python
file_path = './../../data.json'
```

```python
with open(file_path, 'r') as file_in:
    output_json_file = json.load(file_in)

dataset = pd.DataFrame(output_json_file)
```

## Sort data to verify sequentiality

```python
sorted_dataset = dataset.sort_values(by=['file_name', 'el_number'])
```

## Put together all paragraphs by document

```python
grouped_dataset = sorted_dataset[['file_name','content','category']].groupby(['file_name','category'])['content'].apply(lambda x: ' '.join(x)).reset_index()
```

```python
sorted_dataset[sorted_dataset.file_name == '00206BA4E8F5200610123622.pdf']
```

## Analysis category distribution

```python
# Group by category and rename column (with count values) to quantity
grouped_category = grouped_dataset.groupby(['category']).count().reset_index()[['category','file_name']].rename(columns={'file_name': 'quantity'})
```

```python
categories = list(grouped_category['category'])
```

```python
grouped_category
```

```python
print(categories)
```

```python
len(categories)
```

```python
def plot_by_category(dataframe, categories):
    # creating the dataset for plotting
    count = list(dataframe['quantity'])
    fig = plt.figure(figsize = (25, 5))
    # creating the bar plot
    plt.bar(categories, count, color ='green', width = 0.4)
    plt.xlabel("category")
    plt.ylabel("count")
    plt.title("count by category")
    plt.show()
plot_by_category(grouped_category, categories)
```

## Set "others" class for categories with less than 100 samples

```python
joined_dataset = pd.merge(grouped_dataset, grouped_category, how="inner", on=["category"])
```

```python
joined_dataset
```

```python
joined_dataset['updated_category'] = joined_dataset.apply(lambda row: row.iloc[1] if row.iloc[3] >= 100 else 'others' , axis = 1)
```

```python
joined_dataset[joined_dataset.file_name=='vannlekasje hos Sats.pdf']
```

```python
df_train, df_temp = train_test_split(joined_dataset, test_size=0.20, random_state=1, stratify=y)
```
